## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the charity_data.csv.
application_df = pd.read_csv("resources/charity_data.csv")
application_df_copy = application_df.copy()
application_df_copy
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
application_df_copy = application_df_copy.drop(columns = ["EIN", "NAME"])
application_df_copy
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
print(application_df_copy.nunique())
print(application_df_copy.info())
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
application_type_count = application_df_copy["APPLICATION_TYPE"].value_counts()
application_type_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
application_types_to_replace = list(application_type_count[application_type_count < 1100].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df_copy['APPLICATION_TYPE'] = application_df_copy['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df_copy['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
classification_count = application_df_copy["CLASSIFICATION"].value_counts()
classification_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
classification_count_over_2 = classification_count[classification_count > 1]
classification_count_over_2
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
classifications_to_replace = list(classification_count[classification_count < 2000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df_copy['CLASSIFICATION'] = application_df_copy['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df_copy['CLASSIFICATION'].value_counts()
```

```python
print(application_df_copy.nunique())
print(application_df_copy.info())
```

```python
application_df_copy["ORGANIZATION"].value_counts()
```

```python
print(application_df_copy["INCOME_AMT"].value_counts())
print(application_df_copy["INCOME_AMT"].unique())
```

```python
# income_amt_list = [["1-9999", "10000-24999"], ["1M-5M", "5M-10M"], ["10M-50M", "50M+"]]
income_amt_replace_list01 = ["1-9999", "10000-24999"]
income_amt_replace_list02 = ["1M-5M", "5M-10M", "10M-50M", "50M+"]

# Combining 1-9999 with 10000-24999 to make 1-24999
for income_amt in income_amt_replace_list01:
        application_df_copy['INCOME_AMT'] = application_df_copy['INCOME_AMT'].replace(income_amt,"1-24999")

# Combining 1M-5M, 5M-10M, 10M-50M, and 50M+ to make 1M-50M+
for income_amt in income_amt_replace_list02:
        application_df_copy['INCOME_AMT'] = application_df_copy['INCOME_AMT'].replace(income_amt,"1M-50M+")

# Combining 10M-50M with 50M+ to make 10M-50M+
# for income_amt in income_amt_replace_list03:
#         application_df_copy['INCOME_AMT'] = application_df_copy['INCOME_AMT'].replace(income_amt,"10M-50M+")

application_df_copy['INCOME_AMT'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
category_data = pd.get_dummies(application_df_copy, columns = ["APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "INCOME_AMT", "SPECIAL_CONSIDERATIONS"])
category_data
```

```python
category_data.info()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = category_data["IS_SUCCESSFUL"].values
X = category_data.drop(["IS_SUCCESSFUL"], 1).values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 23)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()
number_input_features = len(X_train[0])
# number_input_features = 5
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 40
hidden_nodes_layer3 = 2

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
#  YOUR CODE GOES HERE
# nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adamax", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
